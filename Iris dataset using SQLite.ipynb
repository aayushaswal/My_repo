```python
#import required library
import pandas as pd
import numpy as np
import sqlite3
```

```python
#read iris dataset
data= pd.read_csv("IRIS.csv")
```

```python
#create DB and connection
conn= sqlite3.connect('IRIS.db')
cursor=conn.cursor()
```

```python
#create the table in DB and commit changes
create_table_query="""
create table if not exists IRIS(
Id integer primary key,
sepal_length real,
sepal_width real,
petal_length real,
petal_width	 real,
name text)
"""
cursor.execute(create_table_query)
conn.commit()
```

```python
#checking column of IRIS DB
query="SELECT * FROM IRIS"
df = pd.read_sql_query(query, conn)
df
```

```python
#insert the data in db
for index,row in data.iterrows():
    insert_data_query=f"""
    insert into IRIS (Id,sepal_length,sepal_width,petal_length,petal_width,name) values
    ({row['Id']},{row['sepal_length']},{row['sepal_width']},{row['petal_length']},{row['petal_width']},'{row['name']}')"""
    cursor.execute(insert_data_query)
```

```python
#commit change of insert data
conn.commit()
```
