```python
#import required library
import pandas as pd
import numpy as np
import sqlite3
```

```python
#read iris dataset
data= pd.read_csv("baseball.csv")
data
```

```python
conn=sqlite3.connect('baseball.db')
cursor= conn.cursor()
```

```python
#cursor.execute('DROP TABLE BASEBALL')
```

```python
create_table="""
create table if not exists BASEBALL(
Id integer primary key,
Team text,
League text,
Year INTEGER,
Runs_Scored integer,
Runs_Allowed integer,
Wins integer,
On_Base_Percentage real,
Slugging_Percentage real,
Batting_Average real,
Playoffs BLOB, 
RankSeason integer Default NULL,
RankPlayoffs integer Default NULL,
Games_Played integer,
OOBP real default null,
OSLG real default null
)
"""
```

```python
cursor.execute(create_table)
conn.commit()
```

```python
#INTERROGATING THE TABLE
query1="SELECT * FROM BASEBALL"
df1 = pd.read_sql_query(query1, conn)
df1
```

```python
#total number of runs scored bos in a different year
query1='select Team,Runs_Scored,year from baseball where team="BOS"'
df1= pd.read_sql_query(query1, conn)
df1
```

```python
#team with the lowest number of runs allowed in a given year
query2='select team,min(Runs_Scored) as minimum_run, year from baseball'
df2=pd.read_sql_query(query2,conn)
df2
```

```python
#total number of wins for a specific team across all years
query3='select team, sum(Wins) as total_wins from baseball group by team order by total_wins desc'
df3=pd.read_sql_query(query3,conn)
df3
```

```python
#average batting average for a specific team

query4='select avg(Batting_Average) as avg_batting from baseball where team ="MLN" '
df4=pd.read_sql_query(query4,conn)
df4
```

```python
#team with the highest slugging percentage in year
query5='select team, max(Slugging_Percentage) as highest_slugging, year from baseball group by year '
df5=pd.read_sql_query(query5,conn)
df5
```

```python
#team with the highest number of runs scored in a given league
query5='select team, max(Runs_Scored) as highest_score, year, League  from baseball'
df5=pd.read_sql_query(query5,conn)
df5
```

```python
#team names and the slugging percentages for all teams with a minimum number of games played
query5='select Team ,Slugging_Percentage, min(Games_Played) AS MINIMUM_GAME,YEAR  from BASEBALL GROUP BY YEAR '
df5=pd.read_sql_query(query5,conn)
df5
```

```python
#the team names and playoff status (Yes/No) for all teams in a specific year
query6 = """select team, 
            case when Playoffs=1 then 'Yes'
            else 'No' END AS Playoff_Status from baseball where year=2001"""
df6=pd.read_sql_query(query6,conn)
df6
```

```python
#team with the highest on-base percentage in a given league
query = """SELECT TEAM, MAX(On_Base_Percentage) AS HIGHEST_ON_BASE, LEAGUE FROM BASEBALL GROUP BY League """
df=pd.read_sql_query(query,conn)
df
```

```python
query1= """SELECT Team, SUM(Playoffs) AS PlayoffCount
FROM BASEBALL
GROUP BY Team
HAVING PlayoffCount = (select max(PlayoffCount) from (SELECT Team, SUM(Playoffs) AS PlayoffCount
        FROM baseball
        GROUP BY Team ))"""
df1 = pd.read_sql_query(query1, conn)
df1
```

```python
query1= """SELECT Team, SUM(Playoffs) AS Max_playoff 
FROM BASEBALL
GROUP BY Team order by Max_playoff DESC limit 1
"""
df1 = pd.read_sql_query(query1, conn)
df1
```
